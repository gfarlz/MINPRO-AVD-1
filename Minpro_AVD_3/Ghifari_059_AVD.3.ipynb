## Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Dikarenakan outliers hanya 1 dan berbentuk variabel, maka dilakukan imputasi dengan metode yang paling sesuai, yaitu Winsorizing karena tetap mempertahankan struktur data tanpa membuang nilai ekstrem

```python
columns_to_impute = ["Year"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## Construct data

```python
df['Horsepower'] = pd.to_numeric(df['Horsepower'], errors='coerce')
```

```python
df['0-60 MPH Time (seconds)'] = pd.to_numeric(df['0-60 MPH Time (seconds)'], errors='coerce')
```

```python
df['Combined_Performance'] = (pd.to_numeric(df['Horsepower']) * pd.to_numeric(df['Torque (lb-ft)'])) / pd.to_numeric(df['0-60 MPH Time (seconds)'])

def categorize_horsepower(hp):
    if hp < 300:
        return 'Low'
    elif hp < 500:
        return 'Medium'
    else:
        return 'High'

df['Horsepower_Category'] = df['Horsepower'].apply(categorize_horsepower)

df['Engine_Horsepower_Interaction'] = pd.to_numeric(df['Engine Size (L)']) * pd.to_numeric(df['Horsepower'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/data_bersih.xlsx")
df.head()
```

```python
#Menubah tipe data menjadi numeric
df['Horsepower'] = pd.to_numeric(df['Horsepower'], errors='coerce')
```

```python
# Mengelompokkan tenaga kuda ke dalam rentang tertentu
bin_edges = [100, 200, 300, 400, 500, 600, 700, 800, 1000]
bin_labels = ['100-200 HP', '200-300 HP', '300-400 HP', '400-500 HP',
              '500-600 HP', '600-700 HP', '700-800 HP', '800-1000 HP']

df['Horsepower Category'] = pd.cut(df['Horsepower'], bins=bin_edges, labels=bin_labels, include_lowest=True)
```

```python
hp_by_year = df.groupby('Horsepower Category')['Year'].count()
plt.figure(figsize=(10, 6))
sns.barplot(x=hp_by_year.index, y=hp_by_year.values)
plt.title('Kategori Tenaga Kuda Berdasarkan Tahun')
plt.xlabel('Kategori Tenaga Kuda')
plt.ylabel('Tahun')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi kategori tenaga kuda berdasarkan tahun, terdapat beberapa insight yang bisa diambil. Mayoritas mobil sport dalam dataset memiliki tenaga kuda di rentang **400-700 HP**, dengan puncaknya berada di kategori **400-500 HP dan 600-700 HP**. Sementara itu, mobil sport dengan tenaga kuda sangat rendah (100-200 HP) dan sangat tinggi (800-1000 HP) relatif jarang ditemukan.  

### **Insight yang Dapat Diambil**  
1. **Dominasi Mobil Sport dengan 400-700 HP**  
   - Rentang ini menjadi kategori paling umum dalam industri mobil sport, menunjukkan bahwa mobil dengan tenaga tersebut memiliki keseimbangan yang baik antara performa dan efisiensi.  
   - Konsumen mungkin lebih tertarik pada mobil dalam kisaran ini karena harga yang lebih masuk akal dibandingkan hypercar dengan tenaga ekstrem.  

2. **Mobil dengan Tenaga Ekstrem (800-1000 HP) Masih Jarang**  
   - Produksi mobil dengan tenaga di atas 800 HP masih terbatas, yang kemungkinan besar disebabkan oleh **biaya produksi yang tinggi, pasar yang lebih niche, serta regulasi emisi yang semakin ketat**.

3. **Penurunan Popularitas Mobil dengan Tenaga Rendah (<300 HP)**
