## Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Dikarenakan outliers hanya 1 dan berbentuk variabel, maka dilakukan imputasi dengan metode yang paling sesuai, yaitu Winsorizing karena tetap mempertahankan struktur data tanpa membuang nilai ekstrem

```python
columns_to_impute = ["Year"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

## Construct data

```python
df['Horsepower'] = pd.to_numeric(df['Horsepower'], errors='coerce')
```

```python
df['0-60 MPH Time (seconds)'] = pd.to_numeric(df['0-60 MPH Time (seconds)'], errors='coerce')
```

```python
df['Combined_Performance'] = (pd.to_numeric(df['Horsepower']) * pd.to_numeric(df['Torque (lb-ft)'])) / pd.to_numeric(df['0-60 MPH Time (seconds)'])

def categorize_horsepower(hp):
    if hp < 300:
        return 'Low'
    elif hp < 500:
        return 'Medium'
    else:
        return 'High'

df['Horsepower_Category'] = df['Horsepower'].apply(categorize_horsepower)

df['Engine_Horsepower_Interaction'] = pd.to_numeric(df['Engine Size (L)']) * pd.to_numeric(df['Horsepower'])
```
